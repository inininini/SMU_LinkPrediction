```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

# 1. Download dataset and transform into the matrix forms

## 1-1. Download dataset

```python
import wget
wget.download("https://github.com/montilab/ppi-context/raw/master/data/v_1_00/PPI-Context.txt")
```

## 1-2. Open raw dataset

```python
with open('PPI-Context.txt') as f:
    raw_data = f.readlines()
```

```python
raw_data
```

## 1-3. Transform into the csv format

```python
### the first row has the information on each columns
col_list = raw_data[0].split('\t')
col_list[-1] = col_list[-1].replace('\n','')
col_list
```

```python
### empty dataframe
data_df = pd.DataFrame(columns=col_list)
```

it is the fastest method to append each element to list and transfrom them into columns

```python
d2_list = []
for dump in range(len(col_list)):
    d1_list = []
    d2_list.append(d1_list)
```

```python
for k in range(1, len(raw_data)):
    now_data = raw_data[k].split('\t')
    now_data[-1] = now_data[-1].replace('\n','')

    for col in range(len(now_data)):
        d2_list[col].append(now_data[col])

for colcol in range(len(col_list)):
    col_name_now = col_list[colcol]
    data_df[col_name_now] = d2_list[colcol]   

# data_df.to_csv('context-PPI.csv', index=False)  
```

```python
# data_df = pd.read_csv('context-PPI.csv')
data_df
```

```python
print('the unique number of gene_a : {}, gene_b : {}, cell_line : {}, species : {}'.format(
    len(data_df['gene_a'].unique()), len(data_df['gene_b'].unique()), len(data_df['cell_name'].unique()), len(data_df['cell_species'].unique())))
```

```python
print('total number of genes contained in the relationship : ', len(set(data_df['gene_a'].values.tolist() + data_df['gene_b'].values.tolist())))
```

## 1-4. Generate the Homosapiens' PPI matrix w/o the information on cell lines

```python
### drop the duplicated PPI
data_df_trim = data_df.drop_duplicates(subset=['gene_a','gene_b'])
data_df_trim
```

```python
### We consider only Homosapiens
data_df_trim_homo = data_df_trim[data_df_trim['cell_species']=='Homo sapiens']
data_df_trim_homo
```

```python
data_df_trim_homo.to_csv('context-PPI_final.csv', index=False)
```

```python
mat_df = pd.crosstab(data_df_trim_homo.gene_a, data_df_trim_homo.gene_b)
idx = mat_df.columns.union(mat_df.index)
mat_df = mat_df.reindex(index = idx, columns=idx, fill_value=0)
mat_df
```

```python
print('A sparsity of the marix : ', len(np.where(mat_df.values>0)[0]) / (mat_df.shape[0]**2))
```
